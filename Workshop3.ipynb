```python
# qn2
import matplotlib.pyplot as plt
animals = ["Bear","Dolphin","Whale"]

pop_2017 = [8,150,80]
pop_2018 = [54,77,54]
pop_2019 = [93,32,100]
pop_2020 = [116,11,76]
pop_2021 = [137,6,93]
pop_2022 = [184,1,72]

years = [pop_2017, pop_2018, pop_2019, pop_2020, pop_2021, pop_2022]
year_labels = ["2017", "2018", "2019", "2020", "2021", "2022"]

for i in range(6):
    plt.figure()
    plt.pie(years[i], labels=animals, autopct='%1.1f%%')
    plt.title(f"Animal Population in {year_labels[i]}")
```

```python
#qn2
import matplotlib.pyplot as plt
import numpy as np

animals = ["bears","dolphins","whales"]

y1=[8,150,80]
y2=[54,77,54]
y3=[93,32,100]
y4=[116,11,76]
y5=[137,6,93]
y6=[184,1,72]

fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs = axs.flatten()

data = [y1, y2, y3, y4, y5, y6]
years = ['2017', '2018', '2019', '2020', '2021', '2022']

for i in range(6):
    axs[i].pie(data[i], labels=animals, autopct='%1.1f%%')
    axs[i].set_title(years[i])

plt.legend()
```

```python
import matplotlib.pyplot as plt
CarColor=["Red","Silver","Blue","White","Green"]
data = [123, 230, 78, 193, 12]
c=["red","silver","blue","grey","Green"]
myexplode = [0, 0.2, 0, 0.1,0]
plt.pie(data,labels=CarColor,colors=c,autopct="%0.1f%%",explode = myexplode,shadow = True)
plt.legend(title = "Car Colours:")
plt.show()
```
