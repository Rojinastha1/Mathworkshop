```python
import matplotlib.pyplot as p
groups=['Bears', 'Dolphins', 'Whales']
y1 = [8, 150, 80]
y2 = [54, 77, 54]
y3 = [93, 32, 100]
y4 = [116, 11, 76]
y5 = [137, 6, 93]
y6 = [181, 1 ,71]
years = [y1, y2, y3, y4, y5, y6]
yearsl = [2017, 2018, 2019, 2020, 2021, 2022]
for i in range (6):
  p.subplot(2, 3, i+1)
  p.pie( years[i], labels=groups)
  p.title(yearsl[i])
p.show()
```
