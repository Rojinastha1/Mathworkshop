```python
# Problem 2
import numpy as np
def dot_product(vec1, vec2):
  vec1 = vec1.flatten()
  vec2 = vec2.flatten()
  return np.sum(np.multiply(vec1, vec2))
def matrix_vector_product(mat, vec):
  m,n = mat.shape
  p,q = vec.shape
  assert n == p, "the matrix and vector are not compatible for multiplication!"
  result = np.zeros((m,q), dtype=float)
  for i in range(m):
    result[i] = dot_product(mat[i,:], vec)
  return result
if __name__ == "__main__":
  mat1 = np.full((3,3), 3, dtype=int)
  vec1 = np.full((3,1), 2, dtype=int)
  print("Matrix = ")
  print(mat1)
  print("Vector = ")
  print(vec1)
  result1 = matrix_vector_product(mat1, vec1)
  print("Result = ")
  print(result1)
  result2 = mat1 @ vec1
  print("Result = ")
  print(result2)
```

```python
# Problem 3
import numpy as np
def mat_mat(A, B):
    result = [[0]*len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result
A = [[1,2,3],
     [4,5,6],
     [7,8,9]]
x = [[1],[1],[1]]
print(mat_mat(A,x))
A_np = np.array(A)
x_np = np.array(x)
print(A_np @ x_np)
```

```python
# Problem 4
def is_symmetric(A):
    A = np.array(A)
    return np.array_equal(A, A.T)
A = [[1,2,3],
     [2,5,6],
     [3,6,9]]
print(is_symmetric(A))
```
