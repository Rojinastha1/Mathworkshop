```python
def SumDigits(n):
    s = 0

    while n > 0:
        digit = n % 10
        s = s + digit
        n = n // 10

    print("Sum of digits:", s)

SumDigits(234)
```

```python
def squaredSum(arr):
    total = 0

    for num in arr:
        total = total + (num * num)

    return total

print(squaredSum([2,3,4]))
```

```python
def sumDigits(num):
    total = 0

    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10

    return total

print(sumDigits(678))
```
